```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import pandas as pd
```

```python
alamat = "https://www.kompas.com/"
html = urlopen(alamat)
data = bs(html, 'html.parser')

data
```

```python
table = data.find("div", {"class":"most"})
rows = table.find_all('h4')

rows
```

```python
top_news = [row.get_text().strip() for row in rows]
df_news = pd.DataFrame(top_news, columns=['Judul Berita Populer'])
print(df_news)
```
